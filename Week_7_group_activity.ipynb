### 4. Correlation Analysis

### a. Chloramines vs ph

```python
df1 = df[['ph', 'Chloramines']]
```

```python
corrmatrix = df1.corr()
```

```python
sns.heatmap(corrmatrix, annot=True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.set_theme()
sns.scatterplot(y=df['ph'], x=df['Chloramines'])
plt.title('Chloramines vs ph', fontsize=14)
plt.show()
```

#### Insights:
Ph of pure water ranges from 6.5 to 8.5. From the data, ph values and chloramine levels exhibit only slight negative correlation value of -0.025. It is difficult to discern meaningful relation between both from the scatterplot too.

### b. Conductivity vs ph

```python
df2 = df[['ph', 'Conductivity']]
corrmatrix_2 = df2.corr()
```

```python
sns.heatmap(corrmatrix_2, annot=True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['ph'], y=df['Conductivity'])
plt.show()
```

#### Insights:
Conductivity and ph exhibit only slight positive correlation value of 0.014. As above, no meaninful relation can be inferred from the scatter plot.

```python
full_corrmatrix = df.corr()
```

```python
plt.figure(figsize=(10, 5))
sns.heatmap(full_corrmatrix, annot=True)
plt.show()
```

#### General insights:
From the overall heatmap, parameters doesn't show much correlation. Only ph and sulfates; ph and hardness exhibit comparatively higher correlation but still remains negligible.
